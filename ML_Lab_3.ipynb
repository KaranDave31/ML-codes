```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.arange(0,10)
Y = np.arange(11,21)
```

```python
plt.scatter(X,Y,c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
# plt.savefig('scatter.png')
```

```python
plt.subplot(2,2,1)
plt.plot(X,Y,'g')

plt.subplot(2,2,2)
plt.plot(X,Y,'r--')

plt.subplot(2,2,3)
plt.plot(X,Y,'b*--')

plt.subplot(2,2,4)
plt.plot(X,Y,'go')
```
