```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML labs/iris.csv')
```

```python
df['species'].unique()
```

```python
species = pd.get_dummies(df['species'])
```

```python
species.head(5)
```

```python
df.drop('species', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
df_scaled = std_scaler.fit_transform(df)
```

```python
df_transformed=pd.DataFrame(df_scaled,columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
```

```python
df_transformed
```

```python
iris_df = pd.concat([df_transformed,species],axis=1)
```

```python
iris_df
```

```python
input = iris_df.iloc[:,0:4]
```

```python
input.shape
```

```python
np.random.seed(42)
```

```python
w_i_h1 = np.random.rand(4,1)
w_i_h2 = np.random.rand(4,1)

w_i_o1 = np.random.rand(2,1)
w_i_o2 = np.random.rand(2,1)
w_i_o3 = np.random.rand(2,1)

bias1 = np.random.rand(1)
bias2 = np.random.rand(1)
```

```python
print(w_i_h1)
```

```python
print(w_i_h2)
print(w_i_o1)
print(w_i_o2)
print(w_i_o3)
```

```python
def sigmnoid(x):
  return 1/(1+np.exp(-x))
```

```python
Z2_1 = np.dot(input,w_i_h1) + bias1
Z2_2 = np.dot(input,w_i_h2) + bias2
```

```python
h2_1 = sigmnoid(Z2_1)
h2_2 = sigmnoid(Z2_2)
```

```python
h2_1.shape
```

```python
h2 = np.append(h2_1,h2_2,axis=1)
```

```python
h2.shape
```

```python
Z3_1 = np.dot(h2,w_i_o1)
Z3_2 = np.dot(h2,w_i_o2)
Z3_3 = np.dot(h2,w_i_o3)
```

```python
o1 = sigmnoid(Z3_1)
o2 = sigmnoid(Z3_2)
o3 = sigmnoid(Z3_3)
```

```python
f1 = np.array(iris_df.iloc[:,4])
```

```python
f1.shape
```

```python
f1 = f1.reshape((-1,1))
```

```python
f1.shape
```

```python
e1 = f1-o1
e1.shape
```

```python
f2 = np.array(iris_df.iloc[:,5])
f2 = f2.reshape((-1,1))
e2 = f2-o2
e2.shape
```

```python
f3 = np.array(iris_df.iloc[:,6])
f3 = f3.reshape((-1,1))
e3 = f3-o3
e3.shape
```

```python

error = np.multiply(e1,e1)+np.multiply(e2,e2)+np.multiply(e3,e3)
```

```python
total_error = np.sum(error)
```
