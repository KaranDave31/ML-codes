```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/Salary_Data.csv")
```

```python
xx = df['YearsExperience'].values.reshape(-1,1)
yy = df['Salary'].values.reshape(-1,1)
```

```python
plt.scatter(xx, yy, label='Training Data', color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Years of Experience vs Salary')
plt.legend()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(xx, yy, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
#5.	Implement Gradient Descent with following parameters. Capture intercept, slope for each iteration, and plot a graph between intercept and Mean Squared Error. Also, plot regression line. Check number of iterations it took to complete converge.
#a.	Maximum iterations =1000
#b.	Learning Rate (eta0)=0.01
```

```python
max_iter = 1000
eta0 = 0.01
tolerance = 0.001
```

```python
intercepts = []
mse_values = []
```

```python
model = SGDRegressor(max_iter=1, tol=None, eta0=eta0, random_state=42 ,warm_start=True)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
for i in range(max_iter):
  model.fit(X_train, y_train)
  intercepts.append(model.intercept_[0])
  y_pred_train = model.predict(X_train)
  mse = mean_squared_error(y_train, y_pred_train)
  mse_values.append(mse)

  if i>0 and abs(mse_values[-2] - mse_values[-1])<tolerance:
    print(f"Converged in {i+1} iterations")
    break
```

```python
y_pred_test = model.predict(X_test)
r2 = r2_score(y_test, y_pred_test)
print(f"R^2 Score: {r2}")
```

```python
plt.scatter(X_train, y_train, label='Training Data', color='red')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Years of Experience vs Salary')
plt.legend()
plt.show()
```

```python
plt.scatter(X_test, y_test, label='Test Data', color='pink')
plt.plot(X_test, model.predict(X_test), color='green')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Years of Experience vs Salary')
plt.legend()
plt.show()
```

```python
plt.plot(intercepts, mse_values, marker="o", color="purple")
plt.xlabel("Intercept")
plt.ylabel("Mean Square Error (MSE)")
plt.title("Intercept vs MSE")
plt.grid()
plt.show()
```
